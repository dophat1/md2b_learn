```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
petrol_price_comparison = pd.read_csv("data\\petrol_prices_comparison.csv")
oil_price_daily = pd.read_csv("data\\crude_oil_daily.csv")
```

```python
oil_price_daily
```

```python
fig, ax = plt.subplots(figsize = (13,5))

ax.barh(petrol_price_comparison['Country'], petrol_price_comparison['Pct_Increase'], color=plt.cm.tab20(range(len(petrol_price_comparison['Country']))))
ax.set_xlabel("Country", labelpad=20)
ax.set_ylabel("%")
ax.set_title("Pct of oil price increases")
```

```python
top5 = petrol_price_comparison.sort_values("Amount_Change", ascending=False).head(5)
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(top5['Country'], top5['Amount_Change'], color=plt.cm.tab10(range(5)))
ax.set_xlabel("Country")
ax.set_ylabel("Amount change")
```

```python
oil_price_daily['Date'] = pd.to_datetime(oil_price_daily['Date'])

fig, ax = plt.subplots(figsize=(10,5))
ax.plot(oil_price_daily['Date'], oil_price_daily['Brent_USD'])
ax.plot(oil_price_daily['Date'], oil_price_daily['WTI_USD'])

ax.axvline(pd.Timestamp('03-02-2026'), color="blue", linestyle='--', label="Iran conflict started")
ax.set_xlabel('Date')
ax.set_ylabel('Price')
fig.autofmt_xdate()
ax.legend()
ax.set_title("Oil price changes in Brent USD and WTI USD")

import matplotlib.dates as mdates

ax.xaxis.set_major_locator(mdates.DayLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter("%Y-%m-%d"))
fig.autofmt_xdate()
```
